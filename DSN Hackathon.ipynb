```python
#this function deals with outliers 
def outlier_zscore(df):
    df= df.select_dtypes(include= 'number').copy()
    for col in df.columns:
        mean, std = df[col].mean(), df[col].std()
        upper=  mean+ 3*std  
        lower= mean - 3*std
        df[col][df[col] > upper]= mean
        df[col][df[col] < lower]= mean
    return df
```

```python
#dealt with outliers in dataset
df = outlier_zscore(df)
```

```python
df.info()
```

### Training Model

```python
X = df.drop(['good_bad_flag'], axis = 1)
y = df['good_bad_flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_predict = lr_model.predict(X_test)
```

```python
f1_score(y_test,lr_predict)
```
